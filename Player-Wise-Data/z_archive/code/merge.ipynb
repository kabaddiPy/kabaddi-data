```python
import pandas as pd
import os
import re


def extract_stat_name(filename):
    # Extract the stat name from the filename
    match = filename.split("_")[2:][0].split(".")[0]

    return match


# print(extract_stat_name("df_season9_player-successful-raids.csv"))


def merge_csvs(directory):
    # Get all CSV files in the directory
    csv_files = [f for f in os.listdir(directory) if f.endswith(".csv")]

    # Initialize an empty list to store dataframes
    dfs = []

    for file in csv_files:
        # Read each CSV file
        df = pd.read_csv(os.path.join(directory, file))

        # Ensure 'player_id' is in the dataframe
        if "player_id" not in df.columns:
            print(f"Warning: 'player_id' not found in {file}. Skipping this file.")
            continue

        # Extract the stat name from the filename
        stat_name = extract_stat_name(file)

        # Rename columns to avoid conflicts
        for col in df.columns:
            if col != "player_id":
                df.rename(columns={col: f"{stat_name}_{col}"}, inplace=True)

        dfs.append(df)

    # Merge all dataframes
    merged_df = dfs[0]
    for df in dfs[1:]:
        merged_df = pd.merge(merged_df, df, on="player_id", how="outer")

    return merged_df


# Get directory from user
directory = "./season10"

# Merge CSVs
result = merge_csvs(directory)


# result.head()
```
