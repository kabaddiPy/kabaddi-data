```python
import pandas as pd
import json
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time


import requests
import csv
from bs4 import BeautifulSoup
from selenium  import webdriver
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
import os


d={"s-10": "44", 
   "s-9": "25", 
   "s-8": "20",
   "s-7": "11",
   "s-6": "10",
   "s-5": "8",
   "s-4": "4",
   "s-3": "3",
   "s-2": "2",
   "s-1": "1",
   "s-all":"0" ,
   
   }


def extract_json_from_script(script_content):
    start = script_content.find('window.statsWidgetData = ')
    if start == -1:
        return None
    start += len('window.statsWidgetData = ')
    
    bracket_count = 0
    end = start
    for i, char in enumerate(script_content[start:]):
        if char == '{':
            bracket_count += 1
        elif char == '}':
            bracket_count -= 1
            if bracket_count == 0:
                end = start + i + 1
                break
    
    if bracket_count != 0:
        return None
    
    json_str = script_content[start:end]
    
    try:
        return json.loads(json_str)
    except json.JSONDecodeError:
        return None

def fetch_and_process_data(url):
    options = webdriver.FirefoxOptions()
    options.add_argument("--headless")
    driver = webdriver.Firefox(options=options)
    
    try:
        driver.get(url)
        
        for i in range(21):
            try:
                element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CLASS_NAME, 'loadmore'))
                )
                driver.execute_script("arguments[0].click();", element)
                time.sleep(1)
            except:
                print(f"Stopped after {i} clicks")
                break
        
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        script_tag = soup.find('script', string=lambda text: text and 'window.statsWidgetData' in text)
        
        if script_tag:
            json_data = extract_json_from_script(script_tag.string)
            if json_data:
                player_data = json_data['stats']['data']
                df = pd.DataFrame(player_data)
                
                df['value'] = pd.to_numeric(df['value'])
                df['match_played'] = pd.to_numeric(df['match_played'])
                
                df['points_per_match'] = df['value'] / df['match_played']
                
                return df, json_data['stats']['stat']
    
    finally:
        driver.quit()
    
    return None, None

urls = [
    f"https://www.prokabaddi.com/stats/{d['s-5']}-102-total-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-21-successful-raids-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-22-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-23-successful-tackles-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-103-tackle-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-22-avg-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-139-avg-tackle-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-132-do-or-die-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-104-super-raids-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-28-super-tackles-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-100-super-10s-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-5']}-101-high-5s-statistics"
]   

x = 0
for url in urls:
    x+=1
    df, stat_name = fetch_and_process_data(url)
    if df is not None:
        # Create a file name based on the statistic
        file_name = f"{x}_df_season5_{stat_name.lower().replace(' ', '-')}.csv"
        
        # Export the DataFrame to CSV
        df.to_csv(file_name, index=False)
        print(f"Exported {file_name}")
        
        # Display basic information about the DataFrame
        print(f"\nStatistic: {stat_name}")
        print(df.info())
        
        # Display the first few rows
        print(df.head())
        
        # Basic statistics
        print(df.describe())
        
        # Top 10 players by the statistic
        top_10_players = df.sort_values('value', ascending=False).head(10)
        print(f"\nTop 10 players by {stat_name}:")
        print(top_10_players[['player_name', 'value', 'team_name']])
        
        # Team performance
        team_performance = df.groupby('team_name')['value'].sum().sort_values(ascending=False)
        print(f"\nTeam performance (total {stat_name}):")
        print(team_performance)
        
        # Position-wise analysis
        position_analysis = df.groupby('position_name')['value'].agg(['sum', 'mean', 'count'])
        print("\nPosition-wise analysis:")
        print(position_analysis)
        
        # Correlation between matches played and the statistic
        correlation = df['match_played'].corr(df['value'])
        print(f"\nCorrelation between matches played and {stat_name}: {correlation:.2f}")
        
        print("\n" + "="*50 + "\n")
    else:
        print(f"Failed to fetch or process data from {url}")
```

```python
import pandas as pd
import json
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
import time


import requests
import csv
from bs4 import BeautifulSoup
from selenium  import webdriver
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
import os


d={"s-10": "44", 
   "s-9": "25", 
   "s-8": "20",
   "s-7": "11",
   "s-6": "10",
   "s-5": "8",
   "s-4": "4",
   "s-3": "3",
   "s-2": "2",
   "s-1": "1",
   "s-all":"0" ,
   
   }


def extract_json_from_script(script_content):
    start = script_content.find('window.statsWidgetData = ')
    if start == -1:
        return None
    start += len('window.statsWidgetData = ')
    
    bracket_count = 0
    end = start
    for i, char in enumerate(script_content[start:]):
        if char == '{':
            bracket_count += 1
        elif char == '}':
            bracket_count -= 1
            if bracket_count == 0:
                end = start + i + 1
                break
    
    if bracket_count != 0:
        return None
    
    json_str = script_content[start:end]
    
    try:
        return json.loads(json_str)
    except json.JSONDecodeError:
        return None

def fetch_and_process_data(url):
    options = webdriver.FirefoxOptions()
    options.add_argument("--headless")
    driver = webdriver.Firefox(options=options)
    
    try:
        driver.get(url)
        
        for i in range(21):
            try:
                element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CLASS_NAME, 'loadmore'))
                )
                driver.execute_script("arguments[0].click();", element)
                time.sleep(1)
            except:
                print(f"Stopped after {i} clicks")
                break
        
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        script_tag = soup.find('script', string=lambda text: text and 'window.statsWidgetData' in text)
        
        if script_tag:
            json_data = extract_json_from_script(script_tag.string)
            if json_data:
                player_data = json_data['stats']['data']
                df = pd.DataFrame(player_data)
                
                df['value'] = pd.to_numeric(df['value'])
                df['match_played'] = pd.to_numeric(df['match_played'])
                
                df['points_per_match'] = df['value'] / df['match_played']
                
                return df, json_data['stats']['stat']
    
    finally:
        driver.quit()
    
    return None, None

urls = [
    f"https://www.prokabaddi.com/stats/{d['s-1']}-102-total-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-21-successful-raids-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-22-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-23-successful-tackles-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-103-tackle-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-22-avg-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-139-avg-tackle-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-132-do-or-die-raid-points-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-104-super-raids-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-28-super-tackles-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-100-super-10s-statistics",
    f"https://www.prokabaddi.com/stats/{d['s-1']}-101-high-5s-statistics"
]   

x = 0
for url in urls:
    x+=1
    df, stat_name = fetch_and_process_data(url)
    if df is not None:
        # Create a file name based on the statistic
        file_name = f"{x}_df_season1_{stat_name.lower().replace(' ', '-')}.csv"
        
        # Export the DataFrame to CSV
        df.to_csv(file_name, index=False)
        print(f"Exported {file_name}")
        
        # Display basic information about the DataFrame
        print(f"\nStatistic: {stat_name}")
        print(df.info())
        
        # Display the first few rows
        print(df.head())
        
        # Basic statistics
        print(df.describe())
        
        # Top 10 players by the statistic
        top_10_players = df.sort_values('value', ascending=False).head(10)
        print(f"\nTop 10 players by {stat_name}:")
        print(top_10_players[['player_name', 'value', 'team_name']])
        
        # Team performance
        team_performance = df.groupby('team_name')['value'].sum().sort_values(ascending=False)
        print(f"\nTeam performance (total {stat_name}):")
        print(team_performance)
        
        # Position-wise analysis
        position_analysis = df.groupby('position_name')['value'].agg(['sum', 'mean', 'count'])
        print("\nPosition-wise analysis:")
        print(position_analysis)
        
        # Correlation between matches played and the statistic
        correlation = df['match_played'].corr(df['value'])
        print(f"\nCorrelation between matches played and {stat_name}: {correlation:.2f}")
        
        print("\n" + "="*50 + "\n")
    else:
        print(f"Failed to fetch or process data from {url}")
```
