```python
import pandas as pd
import os

```

```python
files = []
for file in os.listdir("."):
    if file.endswith("conv_touched.csv"):
        files.append(file)

print(len(files), files)

seas = "s7"
```

```python
for file in files:
    print(file.split(seas)[0].strip())
print(f"\nlen files: {len(files)}")
```

```python
df_list = []


for file in files:
    df = pd.read_csv(f"./{file}")
    df['team'] = file.split(seas)[0].strip()
    df['season'] = "PKL-S" + seas[-1]
    df_list.append(df)
    print(df.head())
    print("\n\n")


```

```python
merged_df = pd.concat(df_list, axis=0, ignore_index=True)
```
